```python
data.sort_values(by ='hours_of_sleep', ascending =False)
```

```python
sortedByHourSleept = data.groupby(by=["hours_of_sleep"]).size()
step_count = data.groupby(by=["step_count"]).size()
mood = data.groupby(by=["mood"]).size()
```

```python
sortedByHourSleept
sleepHours = pd.DataFrame({'sleepHours':sortedByHourSleept.index, 'days':sortedByHourSleept.values})
stepDays = pd.DataFrame({'steps':step_count.index, 'days':step_count.values})
moodDays = pd.DataFrame({'mood':mood.index, 'days':mood.values})
```

```python
sleepHours.plot(x ='sleepHours', y='days', kind = 'line', figsize=(20, 10), title ='SleepHoursPerDays')
```

```python
stepDays.plot(x ='steps', y='days', kind = 'scatter', figsize=(20, 10), title ='StepsByDays')
```

```python
moodDays.plot(x ='mood', y='days', kind = 'scatter', figsize=(20, 10), title ='MoodByDays')
```

```python
print(data2.columns)
```

```python
uniqueReasonForNotTraining = data2['What barriers, if any, prevent you from exercising more regularly?           (Please select all that apply)'].unique()
typesOfExercises = data2['What form(s) of exercise do you currently participate in ?                        (Please select all that apply)'].unique()
timeOfTraining = data2['What time if the day do you prefer to exercise?'].unique()
```

```python
reasons = data2.groupby(by=['What barriers, if any, prevent you from exercising more regularly?           (Please select all that apply)']).size()
```

```python
reasonsRepetitions = pd.DataFrame({'reason':reasons.index, 'repetitions':reasons.values})
```

```python
sortedReasons = reasonsRepetitions.sort_values(by='repetitions', ascending = False)
limitedSortedReason = sortedReasons[:9]
```

```python
plot = limitedSortedReason.plot.pie(y='repetitions',labels=limitedSortedReason['reason'], figsize=(15, 15), title="Mainly reason of not exercising")
```

```python
activities = data2.groupby(by=['What form(s) of exercise do you currently participate in ?                        (Please select all that apply)']).size()
activitiesTypes = pd.DataFrame({'activity':activities.index, 'repetitions':activities.values})
```

```python
sortedActivities = activitiesTypes.sort_values(by='repetitions', ascending = False)
limitedSortedActivities = sortedActivities[:9]
```

```python
plot = limitedSortedActivities.plot.pie(y='repetitions',labels=limitedSortedActivities['activity'], figsize=(15, 15), title="Mainly activities types")
```

```python
motivation = data2.groupby(by=['What motivates you to exercise?         (Please select all that applies )']).size()
motivationDf = pd.DataFrame({'motivation':motivation.index, 'repetitions':motivation.values})
```

```python
sorted = motivationDf.sort_values(by='repetitions', ascending=False)
```

```python
plot = limitedSortedActivities.plot.pie(y='repetitions',labels=sorted['motivation'], figsize=(15, 15), title="Mainly motivation")
```

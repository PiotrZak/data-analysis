```python
import numpy as np 

# practically soccer and football is only language issue
df['Sport'][df['Sport'] == 'Soccer'] = 'Football'
df['Sport'][df['Sport'] == 'Basketbal'] = 'Basketball'
df['Sport'][df['Sport'] == 'Mixed Martial Arts'] = 'MMA'
df['Sport'][df['Sport'] == 'Auto Racing'] = 'Racing'

```

```python
basketball = df[df.Sport == 'Basketball']
football = df[df.Sport == 'Football']
```

```python
basketball.plot(x="Year", y="Earnings")
```

```python
football.plot(x="Year", y="Earnings")
```

```python
football
```

```python
footballEarningPerYear = football.groupby(by=["Year"]).sum().sort_values('Year')
```

```python
footballEarningPerYear.sort_values(by='Year', ascending = False)
```
